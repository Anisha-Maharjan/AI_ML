```python
# Task 1
def convert_units():
    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter the value to convert: "))

        if choice == "1":
            print("1. Meters to Feet")
            print("2. Feet to Meters")
            sub = input("Choose conversion: ")

            if sub == "1":
                result = value * 3.28084
                print(f"{value} meters = {result} feet")

            elif sub == "2":
                result = value / 3.28084
                print(f"{value} feet = {result} meters")

        elif choice == "2":
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")
            sub = input("Choose conversion: ")

            if sub == "1":
                result = value * 2.20462
                print(f"{value} kg = {result} lbs")

            elif sub == "2":
                result = value / 2.20462
                print(f"{value} lbs = {result} kg")

        elif choice == "3":
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")
            sub = input("Choose conversion: ")

            if sub == "1":
                result = value * 0.264172
                print(f"{value} L = {result} gallons")

            elif sub == "2":
                result = value / 0.264172
                print(f"{value} gallons = {result} L")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

convert_units()
```

4.3 Exercise on Nested Lists

```python
# 1 Flatten Nested List
def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(i)
        else:
            result.append(i)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

```python
# 2 Access Nested Element
def access_nested_element(lst, indices):
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

```python
# 3 Sum Nested List
def sum_nested(lst):
    total = 0
    for i in lst:
        if isinstance(i,list):
            total += sum_nested(i)
        else:
            total += i
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
# 4 Remove Element
def remove_element(lst, elem):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.append(remove_element(i,elem))
        elif i != elem:
            result.append(i)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
# 5 Find Max
def find_max(lst):
    max_val = float('-inf')

    for i in lst:
        if isinstance(i,list):
            max_val = max(max_val,find_max(i))
        else:
            max_val = max(max_val,i)

    return max_val

print(find_max([[1,2],[3,[4,5]],6]))
```

```python
# 6 Count Occurrences
def count_occurrences(lst,elem):
    count = 0

    for i in lst:
        if isinstance(i,list):
            count += count_occurrences(i,elem)
        elif i == elem:
            count += 1

    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
# 7 Deep Flatten
def deep_flatten(lst):
    result = []

    for i in lst:
        if isinstance(i,list):
            result.extend(deep_flatten(i))
        else:
            result.append(i)

    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
# 8 Nested Average
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

```python
import time

size = 1000000

list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result = [list1[i] + list2[i] for i in range(size)]
print("Python addition:", time.time() - start)

a = np.arange(size)
b = np.arange(size)

start = time.time()
result = a + b
print("NumPy addition:", time.time() - start)
```

```python
size = 200

matrix1 = [[i for i in range(size)] for j in range(size)]
matrix2 = [[i for i in range(size)] for j in range(size)]

result = [[0]*size for _ in range(size)]

start = time.time()

for i in range(size):
    for j in range(size):
        for k in range(size):
            result[i][j] += matrix1[i][k] * matrix2[k][j]

end = time.time()

print("Python matrix multiplication time:", end - start)
```
